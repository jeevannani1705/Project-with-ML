**Splitting the dataset into Training set and Testing set**

```python
#splitting the given data into training data and testing data
from sklearn.model_selection import train_test_split
features_train,features_test,labels_train,labels_test=train_test_split(features,labels,test_size=0.3)
```

**Encoding the variable**

```python
from sklearn import preprocessing
```

```python
# features_trainl encode the target variable
encoder = preprocessing.LabelEncoder()
features_train["country"] = encoder.fit_transform(features_train["country"])
features_train["notes"] = encoder.fit_transform(features_train["notes"])
features_train["lemmatized"] = encoder.fit_transform(features_train["lemmatized"])
features_train["stemmed"] = encoder.fit_transform(features_train["stemmed"])
# valid_ = encoder.fit_transform7
```

```python
features_train
```

```python
features_test
```

```python
# features_trainl encode the target variable
encoder = preprocessing.LabelEncoder()
labels_train["event_type"] = encoder.fit_transform(labels_train["event_type"])
# valid_y = encoder.fit_transform7
```

```python
labels_train
```

```python
labels_test
```

**Training the Naive Bayes model on the Training set**

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(features_train, labels_train)
```

```python
predict_y=classifier.predict(features_train)
predict_y
```

**Training the Random Forest model on the Training set**

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
classifier.fit(features_train, labels_train)
```

```python
predict_y=classifier.predict(features_train)
predict_y
```

**Training the SVM model on the Training set**

```python
from sklearn.svm import SVC
support_vector=SVC(kernel='linear')
support_vector.fit(features_train,labels_train)
```
